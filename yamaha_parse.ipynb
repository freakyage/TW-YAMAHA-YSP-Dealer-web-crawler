```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.options import Options
from urllib.parse import unquote,quote

import time

import xlsxwriter
```

```python
option = Options()

option.add_argument("--disable-infobars")
option.add_argument("start-maximized")
option.add_argument("--disable-extensions")

# Pass the argument 1 to allow and 2 to block
option.add_experimental_option("prefs", {
    "profile.default_content_setting_values.notifications": 2
})
```

```python
driver = webdriver.Chrome(options = option)
```

```python
url = 'https://www.yamaha-motor.com.tw/service.aspx?v=1{}&km=5&classname=area'
```

### 網頁網址

```python
driver.get(url.format(''))
```

```python
privateButton = driver.find_element(By.CLASS_NAME, "agreePrivacy-btn")
privateButton.click()
```

```python
countyList = driver.find_element(By.ID, "County").text.split()
print(countyList)
```

### Excel分頁名

```python
filename = 'yamaha_ysp_all.xlsx'
workBook = xlsxwriter.Workbook(filename)
labels = ["Title","Subtitle","Phone","Product","Maintenance","Color","Polyline","Description"]
```

```python
for sheetName in countyList:
    workSheet = workBook.add_worksheet(sheetName)
    countyUrl = '&County=' + quote(sheetName)
    driver.get(url.format(countyUrl))
    selectTab = driver.find_element(By.CLASS_NAME, "condition-tabs")
    selectArea = selectTab.find_element(By.CSS_SELECTOR, "li")
    selectArea.click()
    searchButton = driver.find_element(By.ID, "searchbtn")
    searchButton.click()
    time.sleep(2)
    #labels = ["name","address","phone","product","maintenance"]
    for label in range(0, len(labels)):
        workSheet.write(0, label, labels[label])
    storeList = driver.find_element(By.ID, "result-list")
    storeName = storeList.find_elements(By.CLASS_NAME, "store")
    print("店家數：{}".format(len(storeName)))
    for store, row in zip(storeName, range(1, len(storeName)+1)):
        textName = store.find_element(By.CSS_SELECTOR, "h3").text.split()[0]
        textAddress = store.find_element(By.CLASS_NAME, "address").text
        textPhone = store.find_element(By.CLASS_NAME, "phone").text
        textProduct = store.find_element(By.CLASS_NAME, "products").find_element(By.CLASS_NAME, "data").text
        textMaintenance = store.find_element(By.CLASS_NAME, "maintenance").find_element(By.CLASS_NAME, "data").text
        print("{}  {}  {}  {}  {}".format(textName, textPhone, textAddress, textProduct, textMaintenance))
        items = [textName, textAddress, textPhone, textProduct, textMaintenance]
        for item in range(0, len(items)):
            workSheet.write(row, item, items[item])
```

```python
workBook.close()
```
